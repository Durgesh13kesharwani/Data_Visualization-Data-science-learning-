```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Online Sales Data.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Plot the distribution of the 'Units Sold' column
plt.figure(figsize=(10, 6))
sns.histplot(df['Units Sold'], kde=True)
plt.title('Distribution of Units Sold')
plt.show()


```

```python
# Plot the relationship between 'Unit Price' and 'Units Sold'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Unit Price', y='Units Sold', data=df)
plt.title('Relationship between Unit Price and Units Sold')
plt.show()
```

```python
# Create a boxplot of 'Unit Price' by 'Product Category'
plt.figure(figsize=(10, 6))
sns.boxplot(x='Product Category', y='Unit Price', data=df)
plt.title('Unit Price by Product Category')
plt.xticks(rotation=45)
plt.show()

```

```python
# Create a heatmap of the correlation matrix
corr_matrix = df[['Units Sold', 'Unit Price', 'Total Revenue']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
